```python
%matplotlib inline
import pandas as pd
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
import datetime

```

```python
#days=    14feb     21mar,
#2018=    180217   ,180324
#2019=    190216   ,190323,
#2020=    200215   ,200328, 
#2021=    210220   ,210327, ......
dates=[180217,180324,190216,190323,200215,200328,210220]
df=pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_210327.txt')
for i in dates:
    temp= pd.read_csv(f"http://web.mta.info/developers/data/nyct/turnstile/turnstile_{i}.txt")
    df= pd.concat([df, temp])
df.head(50)
```

# DATE_TIME

```python
# Take the date and time fields into a single datetime column
df['DATE_TIME'] = pd.to_datetime((df.DATE + ' ' + df.TIME), format='%m/%d/%Y %H:%M:%S')
```

# DROP THE COLUMN

```python
#DROP THE COLUMN
df=df.drop(['LINENAME','DIVISION','TIME','DESC','EXITS'], axis=1,errors="ignore")
```

# DROP DUPLICATE

```python
# Sanity Check to verify that "C/A", "UNIT", "SCP", "STATION", "DATE_TIME" is unique
(df
 .groupby(["C/A", "UNIT", "SCP", "STATION", "DATE_TIME"])
 .ENTRIES.count()
 .reset_index()
 .sort_values("ENTRIES", ascending=False)).head(5)
```

```python
#DROP DUPLICATE
df.drop_duplicates(subset=["C/A", "UNIT", "SCP", "STATION", "DATE_TIME"], inplace=True)
```

```python
#to check 
(df
 .groupby(["C/A", "UNIT", "SCP", "STATION", "DATE_TIME"])
 .ENTRIES.count()
 .reset_index()
 .sort_values("ENTRIES", ascending=False)).head(5)
```

# 

# 

```python
#grab the last reading of the day (located at the top of the group)
turnstiles_daily = (df
                        .groupby(["C/A", "UNIT", "SCP", "STATION", "DATE"],as_index=False)
                        .ENTRIES.first())

#as_index – Set to True, this provides the output with the grouping columns as the index. When set to False, the values are returned in an SQL-like format
```

# 

```python
turnstiles_daily[["PREV_DATE", "PREV_ENTRIES"]] = (turnstiles_daily
                                                       .groupby(["C/A", "UNIT", "SCP", "STATION"])["DATE", "ENTRIES"]
                                                       .apply(lambda grp: grp.shift(1)))
```

```python
# Drop the rows for the earliest date in the df
turnstiles_daily.dropna(subset=["PREV_DATE"], axis=0, inplace=True)
```

```python
def get_daily_counts(row, max_counter):
    counter = row["ENTRIES"] - row["PREV_ENTRIES"]
    if counter < 0:
        # Maybe counter is reversed?
        counter = -counter
    if counter > max_counter:
        # Maybe counter was reset to 0? 
        print(row["ENTRIES"], row["PREV_ENTRIES"])
        counter = min(row["ENTRIES"], row["PREV_ENTRIES"])
    if counter > max_counter:
        # Check it again to make sure we're not still giving a counter that's too big
        return 0
    return counter

# If counter is > 1Million, then the counter might have been reset.  
# Just set it to zero as different counters have different cycle limits
# It'd probably be a good idea to use a number even significantly smaller than 1 million as the limit!
turnstiles_daily["DAILY_ENTRIES"] = turnstiles_daily.apply(get_daily_counts, axis=1, max_counter=1000000)
```

```python
station_daily = turnstiles_daily.groupby(["C/A", "UNIT", "SCP","STATION", "DATE"])[['DAILY_ENTRIES']].sum().reset_index()
```
